```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os 
import cv2

DATADIR = "E:/GrayScale" #directory of dataset (To be changed with original dowloaded dataset)
IMG_SIZE = 256

training_data = []
original_array = []

def create_training_data():
    path = os.path.join(DATADIR, 'REP') #path to coloured image set(Dog is going to be changed to filename inside DATADIR)
    for img in os.listdir(path):
        try:
            color_img = cv2.imread(os.path.join(path, img)) #colored image (LAB)
            labA = cv2.cvtColor(color_img, cv2.COLOR_BGR2LAB)
            label_array = cv2.resize(labA, (IMG_SIZE, IMG_SIZE))
            
            grey_img = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE) #grayscale image
            feature_array = cv2.resize(grey_img, (IMG_SIZE, IMG_SIZE))
            
            actural_img = cv2.imread(os.path.join(path, img)) #coloured image(RGB)
            original_array = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)
            
            training_data.append([feature_array, label_array])
        except Exception as e:
            pass

create_training_data()
```

```python
print(training_data)
```

```python
x = []
y = []
z = []
for features, labels in training_data:
    x.append(features)
    y.append(labels)
x = np.array(x).reshape(-1, IMG_SIZE, IMG_SIZE, 1) # channels = 1 as features is greyscale image   
```

```python
#save our data for reusability
import pickle

pickle_out = open("Grey.pickle","wb")
pickle.dump(x,pickle_out)
pickle_out.close()

pickle_out = open("LAB.pickle","wb")
pickle.dump(y,pickle_out)
pickle_out.close()

'''pickle_out = open("original.pickle","wb")
pickle.dump(z,pickle_out)
pickle_out.close()
'''
```
